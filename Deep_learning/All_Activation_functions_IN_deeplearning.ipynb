```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define activation functions and their derivatives
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z)**2

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return np.where(z > 0, 1, 0)

def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)

def leaky_relu_derivative(z, alpha=0.01):
    return np.where(z > 0, 1, alpha)

def elu(z, alpha=1.0):
    return np.where(z > 0, z, alpha * (np.exp(z) - 1))

def elu_derivative(z, alpha=1.0):
    return np.where(z > 0, 1, alpha * np.exp(z))
```

```python
# Generate values for plotting
z = np.linspace(-5, 5, 400)

# Plot activation functions
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.plot(z, sigmoid(z))
plt.title('Sigmoid Activation')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(z, tanh(z))
plt.title('Tanh Activation')
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(z, relu(z))
plt.title('ReLU Activation')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(z, leaky_relu(z))
plt.title('Leaky ReLU Activation')
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(z, elu(z))
plt.title('ELU Activation')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot derivatives of activation functions
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.plot(z, sigmoid_derivative(z))
plt.title('Sigmoid Derivative')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(z, tanh_derivative(z))
plt.title('Tanh Derivative')
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(z, relu_derivative(z))
plt.title('ReLU Derivative')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(z, leaky_relu_derivative(z))
plt.title('Leaky ReLU Derivative')
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(z, elu_derivative(z))
plt.title('ELU Derivative')
plt.grid(True)

plt.tight_layout()
plt.show()
```
